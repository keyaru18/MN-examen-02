```python
#grafica de la funcion
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    # La funci칩n que queremos resolver: x^3 - 3x^2 + x - 1
    return x**5 + (5.25*x**4) + (4.125*x**3) - (9.125*x**2) - (14.625*x) - 5.625
x = np.linspace(-6, 2, 400)
y = f(x)
plt.plot(x, y, label='f(x)')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.title('Gr치fica de la funci칩n f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.ylim(-50, 50)
plt.grid()
plt.legend()
plt.show()
    
    
```

```python
from scipy.optimize import newton, bisect

def f(x):
    return x**5 + (5.25*x**4) + (4.125*x**3) - (9.125*x**2) - (14.625*x) - 5.625

def f_prime(x):
    # Derivada de la funci칩n
    return 5*x**4 + (21*x**3) + (12.375*x**2) - (18.25*x) - 14.625

raiz_1 = bisect(f, -5, -3)
raiz_2 = newton(f, x0 = -0.5, fprime=f_prime)
raiz_3 = bisect(f, 1, 2)

print(f'Raiz 1 = {raiz_1}')
print(f'Raiz 2 = {raiz_2}')
print(f'Raiz 3 = {raiz_3}')
```
